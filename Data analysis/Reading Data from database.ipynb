```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('chinook.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute('SELECT * FROM employees LIMIT 5;')
```

```python
results = cur.fetchall()
```

```python
results
```

```python
df = pd.DataFrame(results)
```

```python
df.head()
```

```python
cur.close()
conn.close()
```

Usinfg pandas read_sql method

```python
conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql('SELECT * FROM employees', conn)
```

```python
df.head()
```

```python
df.info()
```

```python
df = pd.read_sql('SELECT * FROM employees', conn, index_col= 'EmployeeId' ,parse_dates= ['BirthDate', 'HireDate'])
```

```python
df.info()
```

<h3>Using pandas read_sql_query</h3>

```python
conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql_query('SELECT * FROM employees LIMIT 5', conn)
```

```python
df.head()
```

```python
df = pd.read_sql_query('SELECT * FROM employees', conn , index_col='EmployeeId', parse_dates=['BirthDate', 'HireDate'])
```

```python
df.head()
```

```python
df.info()
```
